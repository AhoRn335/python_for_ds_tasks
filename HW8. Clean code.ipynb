````python
def calculate_variance(nl):
    """
    Calculate the variance of a list of numbers.

    Args:
        nl (list): List of numeric values.

    Returns:
        int: The integer variance of the input list.

    Raises:
        ValueError: If the input list is empty.
    """

    if len(nl) == 0:
        raise ValueError("Input list is empty, variance cannot be calculated.")

    mean = sum(nl) / len(nl)
    sum_squared_diff = 0

    for n in nl:
        squared_diff = (n - mean) ** 2
        sum_squared_diff += squared_diff

    variance = int(sum_squared_diff / len(nl))
    return variance

````
